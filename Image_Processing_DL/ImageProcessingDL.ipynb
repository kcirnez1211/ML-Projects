```python
#resizing the image
from PIL import Image
```

```python
img = Image.open('/content/dog.jpg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('dog_img_resized.jpg')
```

```python
#displaying the image from numpy array
img_res = mpimg.imread('/content/dog_img_resized.jpg')
img_plt = plt.imshow(img_res)
plt.show()
```

```python
#converting RGB to grayscale
import cv2
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
from google.colab.patches import cv2_imshow
```

```python
#displaying the grayscale image
cv2_imshow(grayscale_img)
```

```python
#saving the grayscale image
cv2.imwrite('dog_grayscale_img.jpg',grayscale_img)
```
